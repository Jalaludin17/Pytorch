```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'], inplace=True)
df.head()
```

###Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
#for standard scaling : same size or unit
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
X_test
```

```python
y_train
```

```python
#encoder for output
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

```python
#numpy to pytorch tensors
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor
```

### Model class

```python
import torch.nn as nn

class Model(nn.Module):
  def __init__(self, input_features):
    super().__init__()
    self.layers = nn.Sequential(
        nn.Linear(input_features, 3),
        nn.ReLU(),
        nn.Linear(3, 1),
        nn.Sigmoid()
    )

  def forward(self, features):
    out = self.layers(features)

    return out

  def loss_function(self, y_pred, y):
    epsilon = 1e-7 #to prevent log(0)
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon) #keeps the y_pred between epsilon and 1-epsilon


  #calculate loss : binary cross entropy
    loss = -(y*torch.log(y_pred) + (1-y)*torch.log(1-y_pred))
    return loss.mean()
```

```python
learning_rate = 0.1
epochs = 25
```

```python
#Training Model
model = Model(X_train_tensor.shape[1])

# Convert tensors to float32
X_train_tensor = X_train_tensor.float()
X_test_tensor = X_test_tensor.float()
y_train_tensor = y_train_tensor.float() # also convert target tensor

#define loops
for epoch in range(epochs):
  #forward pass
  y_pred = model(X_train_tensor)

  #calculate loss
  loss = model.loss_function(y_pred, y_train_tensor)

  #backward pass
  loss.backward()

  #Parameters, weights update
  with torch.no_grad(): #turning off gradient tracking
    model.layers[0].weight -= learning_rate * model.layers[0].weight.grad
    model.layers[0].bias -= learning_rate * model.layers[0].bias.grad

    model.layers[2].weight -= learning_rate * model.layers[2].weight.grad
    model.layers[2].bias -= learning_rate * model.layers[2].bias.grad

  #parameters update
  model.layers[0].weight.grad.zero_()
  model.layers[0].bias.grad.zero_()

  model.layers[2].weight.grad.zero_()
  model.layers[2].bias.grad.zero_()

  #print loss in each epoch
  print(f"Epoch {epoch+1} loss: {loss.item()}")
```

### Evaluation

```python
with torch.no_grad():
  y_pred = model(X_test_tensor)
  y_pred = (y_pred > 0.5).float()

  print(y_pred)
```

```python
with torch.no_grad():
  y_pred = model(X_test_tensor)
  y_pred = (y_pred > 0.5).float()
  accuracy= (y_pred == y_test_tensor).float().mean()
  print(f"Accuracy: {accuracy}")
```
