### Neural Network Module

#### 1.Single Neuron

```python
#create a nn module
import torch
import torch.nn as nn

class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1) #num_features are inputs and 1 is output
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

#### 2.Multi Layer Perceptron

```python
class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.linear1= nn.Linear(num_features, 3)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear1(features)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)
    return out
```

```python
class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.network=nn.Sequential(
        nn.Linear(num_features, 3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid()
    )

  def forward(self, features):
    out = self.network(features)

    return out
```

### New Training Pipeline

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'], inplace=True)
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
#for standard scaling : same size or unit
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
y_train
```

```python
#encoder for output
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
#numpy to pytorch tensors
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor.to(torch.float32)
X_test_tensor.to(torch.float32)
y_train_tensor.to(torch.float32)
y_test_tensor.to(torch.float32)
```

#### 1. Single neuron model 2 (nn modul)

```python
import torch
import torch.nn as nn

class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1) #num_features are inputs and 1 is output
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out

  def loss_function(self, y_pred, y):
    epsilon = 1e-7 #to prevent log(0)
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon) #keeps the y_pred between epsilon and 1-epsilon


  #calculate loss : binary cross entropy
    loss = -(y*torch.log(y_pred) + (1-y)*torch.log(1-y_pred))
    return loss.mean()
```

```python
learning_rate = 0.1
epochs = 25
```

```python
#creat model
model = Model(X_train_tensor.shape[1])

for epoch in range(epochs):
  # Forward pass
  y_pred = model(X_train_tensor)

  loss = model.loss_function(y_pred, y_train_tensor)

  # Backward pass
  loss.backward()

  # Parameters update
  with torch.no_grad():
    model.linear.weight -= learning_rate * model.linear.weight.grad
    model.linear.bias -= learning_rate * model.linear.bias.grad

  # Zero gradients
  model.linear.weight.grad.zero_()
  model.linear.bias.grad.zero_()

  # print loss for monitoring
  print(f"epoch: {epoch}, loss: {loss.item()}")
```

```python
model.linear.weight
```

```python
#Evaluation
with torch.no_grad():
  X_test_tensor = X_test_tensor.float() # Cast X_test_tensor to float32
  y_pred = model(X_test_tensor)
  y_pred = (y_pred > 0.5).float() # Changed threshold to 0.5 for binary classification
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f"Accuracy: {accuracy}")
```

#### 2. Single neuron model 3 (Built in loss function)

```python
import torch
import torch.nn as nn

class Model2(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1) #num_features are inputs and 1 is output
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

```python
learning_rate = 0.1
epochs = 25
loss_function = nn.BCELoss()
```

```python
#creat model
model = Model2(X_train_tensor.shape[1])

for epoch in range(epochs):
  # Forward pass
  y_pred = model(X_train_tensor)

  # Reshape y_train_tensor to match y_pred
  #loss = loss_function(y_pred, y_train_tensor.unsqueeze(1))
  loss = loss_function(y_pred, y_train_tensor.view(-1,1))

  # Backward pass
  loss.backward()

  # Parameters update
  with torch.no_grad():
    model.linear.weight -= learning_rate * model.linear.weight.grad
    model.linear.bias -= learning_rate * model.linear.bias.grad

  # Zero gradients
  model.linear.weight.grad.zero_()
  model.linear.bias.grad.zero_()

  # print loss for monitoring
  print(f"epoch: {epoch}, loss: {loss.item()}")
```

```python
#Evaluation
with torch.no_grad():
  X_test_tensor = X_test_tensor.float() # Cast X_test_tensor to float32
  y_pred = model(X_test_tensor)
  y_pred = (y_pred > 0.5).float() # Changed threshold to 0.5 for binary classification
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f"Accuracy: {accuracy}")
```

#### 3. Single neuron model 4 (Builtin Optimizer)

```python
import torch
import torch.nn as nn

class Model2(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1) #num_features are inputs and 1 is output
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

```python
learning_rate = 0.1
epochs = 25
loss_function = nn.BCELoss()
```

```python
#creat model
model = Model2(X_train_tensor.shape[1])

#optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

for epoch in range(epochs):

  # Forward pass
  y_pred = model(X_train_tensor)

  # Reshape y_train_tensor to match y_pred
  #loss = loss_function(y_pred, y_train_tensor.unsqueeze(1))
  loss = loss_function(y_pred, y_train_tensor.view(-1,1))

  #clear gradients
  optimizer.zero_grad()

  # Backward pass
  loss.backward()

  # Parameters update
  optimizer.step()

  # print loss for monitoring
  print(f"epoch: {epoch}, loss: {loss.item()}")
```
